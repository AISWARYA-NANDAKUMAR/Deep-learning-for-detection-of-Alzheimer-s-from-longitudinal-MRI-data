```python
df = df.drop(['MRI ID','Visit','Hand','Subject ID','EDUC','SES'], axis='columns') # Drop unnecessary columns
```

```python
df['Group'] = df['Group'].replace(['Converted'], ['Demented']) # Target variable
df['Group'] = df['Group'].replace(['Demented', 'Nondemented'], [1,0]) # Target variable
df
```

```python
# bar drawing function
import matplotlib.pyplot as plt
Demented = df[df['Group']==1].value_counts()
Nondemented = df[df['Group']==0].value_counts()
print(len(Demented))
print(len(Nondemented))
```

```python
#Avoid Class Imbalance problem
SAMPLE_SIZE = 183

# take a sample of the majority class 0 (total = 198738)
Demented = df[df['Group']==1].sample(SAMPLE_SIZE)
# take a sample of class 1 (total = 78786)
NonDemented = df[df['Group']==0].sample(SAMPLE_SIZE)

# concat the two dataframes
df = pd.concat([Demented, NonDemented], axis=0).reset_index(drop=True)

# Check the new class distribution
df['Group'].value_counts()
```

```python
y = df['Group'] # Target variable
cols = ['MR Delay','Age','MMSE','CDR','eTIV','nWBV','ASF']
X = df[cols]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 20% test
print('X_train',X_train)
print('X_test',X_test)
print('y_train',y_train)
print('y_test',y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
def Model(X,y,xtest,ytest):
  clf1=LogisticRegression()
  clf2=GaussianNB()
  clf3=RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               ccp_alpha=0.0,
                               class_weight=None,
                               criterion='gini',
                               max_depth=None,
                               max_leaf_nodes=None,
                               max_samples=None,
                               min_impurity_decrease=0.0,
                               min_impurity_split=None,
                               min_samples_leaf=1,
                               min_samples_split=2,
                               min_weight_fraction_leaf=0.0,
                               n_jobs=None,
                               oob_score=False,
                               random_state=None,
                               verbose=0,
                               warm_start=False,
                              max_features = 'sqrt')
  clf4=KNeighborsClassifier(n_neighbors=13)
  clf=[clf1,clf2,clf3,clf4]
  name=['Logisticregression','GaussianNB','Random Forest','KNN']
  model={}
  y_pred={}
  for i in range(len(clf)):
         model['model' + str(i)]=clf[i].fit(X,y)
         print( model['model' + str(i)])
         y_pred['model' + str(i)] = model.get('model' + str(i)).predict(xtest)
         print(y_pred['model' + str(i)])
         score = model.get('model' + str(i)).score(xtest, ytest)
         print(name[i])
         print("Accuracy",score)
         print(confusion_matrix(ytest,y_pred.get(('model' + str(i)))))
         print(classification_report(ytest,y_pred.get(('model' + str(i)))))
         score=0
Model(X_train,y_train,X_test,y_test)
```
